```python
img = cv2.imread(image_path + "lena.jpg")              # 彩色讀取
cv2_imshow(img)        # 顯示修改前影像img
print(f"修改前img[160,150] = {img[160,150]}")
print(f"修改前img[170,150] = {img[170,150]}")
print(f"修改前img[180,150] = {img[180,150]}")
# 紫色長條
for y in range(160,170):                # 修改影像
    for x in range(150,230):
        img[y,x] = [255,0,255]          # 紫色取代
# 白色長條
for z in range(170,180):                # 修改影像:一次一個通道值
    for y in range(150,230):
        for x in range(0,3):            # 一次一個通道值
            img[z,y,x] = 255            # 白色取代
# 黃色長條
for y in range(180,190):                # 修改影像
    for x in range(150,230):
        img[y,x] = [0,255,255]          # 黃色取代
cv2_imshow(img)         # 顯示修改後影像img
print(f"修改後img[160,150] = {img[160,150]}")
print(f"修改後img[170,150] = {img[170,150]}")
print(f"修改後img[180,150] = {img[180,150]}")
```

```python
import cv2

img = cv2.imread(image_path + "lena.jpg")              # 彩色讀取
cv2.imshow("Before modify", img)        # 顯示修改前影像img
print(f"修改前img[160,150,1] = {img.item(160,150,1)}")
print(f"修改前img[170,150,1] = {img.item(170,150,1)}")
print(f"修改前img[180,150,1] = {img.item(180,150,1)}")
# 白色長條
for z in range(160,180):                # 修改影像:一次一個通道值
    for y in range(150,230):
        for x in range(0,3):            # 一次一個通道值
            img.itemset((z,y,x),255)    # 白色取代
cv2.imshow("After modify", img)         # 顯示修改後影像img
print(f"修改後img[160,150,1] = {img.item(160,150,1)}")
print(f"修改後img[170,150,1] = {img.item(170,150,1)}")
print(f"修改後img[180,150,1] = {img.item(180,150,1)}")
cv2.waitKey(0)
cv2.destroyAllWindows()                 # 刪除所有視窗
```
