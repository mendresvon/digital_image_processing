```python
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont
from google.colab.patches import cv2_imshow

def cv2_English_Text(img, text, left, top, textColor, fontSize):
    # 轉為 PIL 格式
    if (isinstance(img, np.ndarray)):
        img = Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

    draw = ImageDraw.Draw(img)

    # Colab 預設的 Linux 字體路徑
    font_path = "/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf"
    try:
        fontText = ImageFont.truetype(font_path, fontSize)
    except:
        fontText = ImageFont.load_default() # 找不到就用系統預設

    # 繪製英文
    draw.text((left, top), text, fill=textColor, font=fontText)

    # 轉回 OpenCV 格式
    return cv2.cvtColor(np.asarray(img), cv2.COLOR_RGB2BGR)

# 執行
img = cv2.imread(image_path + "antarctic.jpg")
if img is not None:
    img = cv2_English_Text(img, "I'm in Antarctica", 220, 100, (255, 0, 0), 50)
    cv2_imshow(img)
else:
    print("找不到圖片，請檢查 image_path 是否正確")
```

```python
from google.colab import output
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# 建立一張空白圖片並存檔
image = np.ones((200, 300, 3), np.uint8) * 255
cv2.imwrite('temp.jpg', image)

# JavaScript 程式碼：顯示圖片並監聽點擊
from IPython.display import display, HTML

def get_click_coordinates():
    display(HTML('''
        <canvas id="canvas" width="300" height="200" style="border:1px solid #000;"></canvas>
        <p id="output">請點擊畫布...</p>
        <script>
            var canvas = document.getElementById('canvas');
            var ctx = canvas.getContext('2d');
            var img = new Image();
            img.src = 'temp.jpg';
            img.onload = function() { ctx.drawImage(img, 0, 0); };

            canvas.onclick = function(e) {
                var rect = canvas.getBoundingClientRect();
                var x = e.clientX - rect.left;
                var y = e.clientY - rect.top;
                document.getElementById('output').innerHTML = '點擊位置: x=' + x + ', y=' + y;
                google.colab.kernel.invokeFunction('notebook.OnMouseAction', [x, y], {});
            };
        </script>
    '''))

# 定義 Python 回呼函數
def OnMouseAction(x, y):
    print(f"Python 收到座標: x={int(x)}, y={int(y)}")

# 註冊回呼函數讓 JavaScript 調用
output.register_callback('notebook.OnMouseAction', OnMouseAction)

# 執行
get_click_coordinates()
```

```python
from google.colab import output
from IPython.display import HTML, display
import numpy as np

# 定義畫布尺寸
width, height = 600, 400

def draw_canvas():
    # 使用 HTML/JavaScript 建立互動介面
    js_code = f'''
    <canvas id="myCanvas" width="{width}" height="{height}" style="border:1px solid #000; cursor:crosshair;"></canvas>
    <p>左鍵：實心圓 | 右鍵：空心圓 | Q：結束</p>

    <script>
    var canvas = document.getElementById("myCanvas");
    var ctx = canvas.getContext("2d");

    // 初始化背景白色
    ctx.fillStyle = "white";
    ctx.fillRect(0, 0, canvas.width, canvas.height);

    canvas.addEventListener("mousedown", function(e) {{
        var rect = canvas.getBoundingClientRect();
        var x = e.clientX - rect.left;
        var y = e.clientY - rect.top;

        // 隨機顏色與半徑
        var r = Math.floor(Math.random() * 40) + 10;
        var color = "rgb("+Math.floor(Math.random()*256)+","
                        +Math.floor(Math.random()*256)+","
                        +Math.floor(Math.random()*256)+")";

        ctx.beginPath();
        ctx.arc(x, y, r, 0, 2 * Math.PI);

        if (e.button === 0) {{ // 左鍵
            ctx.fillStyle = color;
            ctx.fill();
        }} else if (e.button === 2) {{ // 右鍵
            ctx.strokeStyle = color;
            ctx.lineWidth = 3;
            ctx.stroke();
        }}
    }});

    // 禁用右鍵選單以免干擾
    canvas.oncontextmenu = function(e) {{ e.preventDefault(); }};
    </script>
    '''
    display(HTML(js_code))

draw_canvas()
```

```python
from google.colab import output
from IPython.display import HTML, display

def start_interactive_draw():
    # 使用 HTML5 Canvas 與 JavaScript 處理滑鼠與按鍵事件
    html_code = '''
    <div style="background-color: #f0f0f0; padding: 10px; border-radius: 5px;">
        <canvas id="drawCanvas" width="600" height="400" style="border:2px solid #333; background-color: white; cursor: crosshair;"></canvas>
        <div style="margin-top: 10px; font-family: sans-serif;">
            <p><strong>操作說明：</strong></p>
            <ul>
                <li><b>滑鼠左鍵：</b> 畫圓形</li>
                <li><b>滑鼠右鍵：</b> 畫矩形</li>
                <li><b>按住鍵盤 's' 鍵：</b> 畫【實心】，放開則為【空心】</li>
                <li><b>點擊下方按鈕：</b> 清除畫布</li>
            </ul>
            <button onclick="clearCanvas()">清除畫布</button>
        </div>
    </div>

    <script>
    var canvas = document.getElementById('drawCanvas');
    var ctx = canvas.getContext('2d');
    var isSolid = false;

    // 監聽鍵盤 's' 鍵
    window.addEventListener('keydown', function(e) {
        if (e.key.toLowerCase() === 's') isSolid = true;
    });
    window.addEventListener('keyup', function(e) {
        if (e.key.toLowerCase() === 's') isSolid = false;
    });

    function getRandomColor() {
        return 'rgb(' + Math.floor(Math.random()*256) + ',' +
                       Math.floor(Math.random()*256) + ',' +
                       Math.floor(Math.random()*256) + ')';
    }

    canvas.addEventListener('mousedown', function(e) {
        var rect = canvas.getBoundingClientRect();
        var x = e.clientX - rect.left;
        var y = e.clientY - rect.top;
        var color = getRandomColor();

        if (e.button === 0) { // 左鍵：畫圓
            var r = Math.floor(Math.random() * 40) + 10;
            ctx.beginPath();
            ctx.arc(x, y, r, 0, 2 * Math.PI);
            if (isSolid) {
                ctx.fillStyle = color;
                ctx.fill();
            } else {
                ctx.strokeStyle = color;
                ctx.lineWidth = 3;
                ctx.stroke();
            }
        }
        else if (e.button === 2) { // 右鍵：畫矩形
            var pw = Math.floor(Math.random() * 90) + 10;
            var ph = Math.floor(Math.random() * 90) + 10;
            if (isSolid) {
                ctx.fillStyle = color;
                ctx.fillRect(x, y, pw, ph);
            } else {
                ctx.strokeStyle = color;
                ctx.lineWidth = 3;
                ctx.strokeRect(x, y, pw, ph);
            }
        }
    });

    // 禁用右鍵選單
    canvas.oncontextmenu = function(e) { e.preventDefault(); };

    function clearCanvas() {
        ctx.clearRect(0, 0, canvas.width, canvas.height);
        ctx.fillStyle = "white";
        ctx.fillRect(0, 0, canvas.width, canvas.height);
    }
    </script>
    '''
    display(HTML(html_code))

# 啟動互動畫布
start_interactive_draw()
```

```python
from google.colab import output
from IPython.display import HTML, display

def interactive_draw_with_trackbar():
    html_code = '''
    <div style="background-color: #f0f0f0; padding: 15px; border-radius: 10px; width: 620px; font-family: sans-serif;">
        <canvas id="paintCanvas" width="600" height="400" style="border:2px solid #333; background-color: white; cursor: crosshair;"></canvas>

        <div style="margin-top: 15px;">
            <label><b>Thickness 控制 (模式切換):</b></label><br>
            <input type="range" id="modeSlider" min="0" max="1" value="0" style="width: 200px;">
            <span id="modeText">模式：實心 (-1)</span>
        </div>

        <p style="font-size: 0.9em; color: #666;">操作：點擊畫布畫圓。滑動條 0 為實心，1 為線寬 3。</p>
        <button onclick="ctx.clearRect(0,0,600,400); ctx.fillStyle='white'; ctx.fillRect(0,0,600,400);">清除畫布</button>
    </div>

    <script>
    var canvas = document.getElementById('paintCanvas');
    var ctx = canvas.getContext('2d');
    var slider = document.getElementById('modeSlider');
    var modeText = document.getElementById('modeText');

    // 初始化畫布
    ctx.fillStyle = "white";
    ctx.fillRect(0, 0, 600, 400);

    // 更新滑動條文字顯示
    slider.oninput = function() {
        if (this.value == "0") {
            modeText.innerText = "模式：實心 (-1)";
        } else {
            modeText.innerText = "模式：線寬 (3)";
        }
    };

    canvas.addEventListener('mousedown', function(e) {
        var rect = canvas.getBoundingClientRect();
        var x = e.clientX - rect.left;
        var y = e.clientY - rect.top;

        // 隨機顏色與半徑 (10-50)
        var r = Math.floor(Math.random() * 41) + 10;
        var color = 'rgb(' + Math.floor(Math.random()*256) + ',' +
                             Math.floor(Math.random()*256) + ',' +
                             Math.floor(Math.random()*256) + ')';

        ctx.beginPath();
        ctx.arc(x, y, r, 0, 2 * Math.PI);

        if (slider.value == "0") {
            ctx.fillStyle = color;
            ctx.fill();
        } else {
            ctx.strokeStyle = color;
            ctx.lineWidth = 3;
            ctx.stroke();
        }
    });
    </script>
    '''
    display(HTML(html_code))

interactive_draw_with_trackbar()
```
