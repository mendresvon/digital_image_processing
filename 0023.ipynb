```python
from google.colab.patches import cv2_imshow #要在Colab使用cv2顯示影像，一定要先宣告這行
```

```python
import cv2
import numpy as np

height = 160                # 影像高
width = 280                 # 影像寬
# 建立GRAY影像陣列
image = np.zeros((height, width), np.uint8)
image[40:120, 70:210] = 255 # 高在40至120之間,寬在70至210之間,設為255
cv2_imshow(image)  # 顯示影像

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

height = 160                # 影像高
width = 280                 # 影像寬
# 建立GRAY影像陣列
image = np.zeros((height, width), np.uint8)
for y in range(0, height, 20):
    image[y:y+10, :] = 255  # 白色厚度是10
cv2_imshow(image)  # 顯示影像

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

height = 160                # 影像高
width = 280                 # 影像寬
# 使用random.randint()建立GRAY影像陣列
image = np.random.randint(256,size=[height, width],dtype=np.uint8)
cv2_imshow(image)  # 顯示影像

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

height = 160                # 影像高
width = 280                 # 影像寬

# 建立黑色影像陣列
image = np.zeros((height, width), np.uint8)

# 繪製兩個眼睛（白色方框）
image[40:70, 90:130] = 255  # 左眼
image[40:70, 150:190] = 255 # 右眼

# 繪製嘴巴（白色方框）
image[100:120, 100:180] = 255 # 嘴巴

# 顯示影像
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
